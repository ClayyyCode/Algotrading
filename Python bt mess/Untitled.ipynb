```python
data= pd.read_csv('data/QQQ24032022.csv')
```

```python
data['Log returns'] = np.log(data['close']/data['close'].shift())
```

```python
data['Log returns'].std()
data['volatility'] = data['Log returns'].rolling(30).std()*(252**0.5)
```
